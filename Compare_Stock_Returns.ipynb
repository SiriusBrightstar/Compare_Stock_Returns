```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import seaborn as sns
import yfinance as yf

%matplotlib inline
```

```python
yf.pdr_override()

Stock_1 = '^CRSLDX'    #Change: Add Ticker 1 from Yahoo Finance
Stock_2 = '^GSPC'      #Change: Add Ticker 2 from Yahoo Finance

A = pdr.get_data_yahoo(Stock_1, data_source = 'yahoo', start = '2020-1-1', end = '2021-1-1')['Adj Close']
B = pdr.get_data_yahoo(Stock_2, data_source = 'yahoo', start = '2020-1-1', end = '2021-1-1')['Adj Close']

#End part can be removed.
```

```python
Plot = pd.DataFrame()
Plot['Nifty500'] = (A / A.iloc[0] * 100 - 100)      #Change inside [] to what you are comparing
Plot['S&P500'] = (B / B.iloc[0] * 100 - 100)        #Change inside [] to what you are comparing
```

```python
#This Shows the returns from Start date to End Date

Plot
```

```python
#Use this to get returns from Start date to a Particular Date

Plot.loc['2020-01-07']      #Change: Date
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(data=Plot, palette='icefire')
plt.legend(fontsize='x-large', title_fontsize='40', loc = 'upper left')
plt.title('Comparison: Nifty500 vs S&P500', fontsize='x-large')      #Change: Title of Plot
plt.xlabel('Time', fontsize='x-large')
plt.ylabel('Returns in %', fontsize='x-large')
plt.grid(axis = 'y')
```
